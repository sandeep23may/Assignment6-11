```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
dF_titan= pd.read_csv(path)
```

```python
dF_titan.head(5)
dF_titan_by_sex=dF_titan.groupby(by = ('sex'))
```

```python
%matplotlib inline
#1. Create a pie chart presenting the male/female proportion

plt.pie([dF_titan_by_sex.get_group('female').shape[0],dF_titan_by_sex.get_group('male').shape[0]],
         labels=['female','male'],colors=['yellowgreen','lightskyblue'],shadow=True,autopct='%1.1f%%',explode=(0.1,0))
plt.title("Pie chart of male/female proportion")
plt.show()
```

```python
####2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

plt.scatter(dF_titan_by_sex.get_group('female').age,dF_titan_by_sex.get_group('female').fare,label='female',c='yellowgreen')
plt.scatter(dF_titan_by_sex.get_group('male').age,dF_titan_by_sex.get_group('male').fare,label='male',c='lightskyblue')
plt.xlabel('Age')
plt.ylabel('Fare')
#plt.figure(figsize=(16,9))
plt.show()
```

```python
####2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, axisbg="1.0")
ax.scatter(dF_titan_by_sex.get_group('female').age,dF_titan_by_sex.get_group('female').fare, alpha=0.8, c='yellowgreen', edgecolors='none', s=30, label='female')
ax.scatter(dF_titan_by_sex.get_group('male').age,dF_titan_by_sex.get_group('male').fare, alpha=0.8, c='lightskyblue', edgecolors='none', s=30, label='male')

plt.title('Matplot scatter plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(loc=2)
plt.show()
```
